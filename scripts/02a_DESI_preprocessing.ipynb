```python
# opening the file in read mode
# This will convert desi data into a better input file for rest of pipeline
### this step takes a long time to run

files = [file for file in os.listdir('raw_DESI') if "_TL.txt" in file]

def create(file):
    print(file)
    my_file = open('raw_DESI/' + file,"r")
    
    # reading the file
    data = my_file.read()
    
    data_into_list = data.split("\n")
    my_file.close()
    
    for i in range(len(data_into_list)):
        data_into_list[i] = data_into_list[i].split("\t")
    
    pop_list = []
    for i in range(len(data_into_list)):
        if data_into_list[i] == ['']:
            pop_list.append(i)
               
    pop_list.reverse()
    for i in pop_list:
        data_into_list.pop(i)
    
    data_into_list.pop(0)
    data_into_list.pop(0)
    
    for i in range(2,len(data_into_list)):
        vec_length = len(data_into_list[i])
        data_into_list[i].pop()
        data_into_list[i].pop()
        data_into_list[i].pop(0)
    
    del data_into_list[0][0:2]
    del data_into_list[1][0:2]
    
    ion_df = pd.DataFrame({'mass': data_into_list[0], 'rt':data_into_list[1]})
    new_df = pd.DataFrame(columns=['V'+str(i+1) for i in range(len(data_into_list[len(data_into_list[2])]))], data=data_into_list[2:len(data_into_list)])
    
    for i in range(new_df.shape[1]):
        new_df[new_df.columns[i]] = new_df[new_df.columns[i]].astype(float)
        
    
    new_df.rename(columns = {"V1": "X","V2":"Y"}, 
              inplace = True)
    
    new_df.to_csv('raw_DESI/'+file.replace(".txt","_mat.csv"))
    ion_df.to_csv('raw_DESI/'+file.replace(".txt","_ions_list.csv"))

[create(file) for file in files]

#First two items are m/z and time
```
