```python
import os
import pandas as pd

# Choose one mode at a time: "neg1", "neg2", or "pos"
mode = "neg1"

project_root = r"C:/Users/darto/Documents/William/Cornell/Project/Spatial_Registration/Proj02/MPI_Registration"
input_dir = os.path.join(project_root, "DESI_files", "Pixel files annotated", "Pixel_text_files")
output_dir = os.path.join(project_root, "results", "desi_output", mode)

os.makedirs(output_dir, exist_ok=True)

# Keep only files matching the selected mode
files = [
    file for file in os.listdir(input_dir)
    if file.endswith(f"_{mode}.txt")
]

print(f"Found {len(files)} file(s) for mode = {mode}")
for f in files:
    print(f)


def create(file_name):
    print(f"\nProcessing: {file_name}")
    input_path = os.path.join(input_dir, file_name)

    with open(input_path, "r", encoding="utf-8", errors="ignore") as my_file:
        data = my_file.read()

    data_into_list = data.split("\n")

    for i in range(len(data_into_list)):
        data_into_list[i] = data_into_list[i].split("\t")

    # Remove empty rows
    pop_list = []
    for i in range(len(data_into_list)):
        if data_into_list[i] == ['']:
            pop_list.append(i)

    pop_list.reverse()
    for i in pop_list:
        data_into_list.pop(i)

    # Remove first two header lines
    data_into_list.pop(0)
    data_into_list.pop(0)

    # Clean matrix rows
    for i in range(2, len(data_into_list)):
        data_into_list[i].pop()      # remove last column
        data_into_list[i].pop()      # remove second-last column
        data_into_list[i].pop(0)     # remove first column

    # Remove first two entries from m/z and rt rows
    del data_into_list[0][0:2]
    del data_into_list[1][0:2]

    # Create ions metadata table
    ion_df = pd.DataFrame({
        'mass': data_into_list[0],
        'rt': data_into_list[1]
    })

    # Create matrix table
    matrix_data = data_into_list[2:]
    n_cols = len(matrix_data[0])

    new_df = pd.DataFrame(
        columns=[f'V{i+1}' for i in range(n_cols)],
        data=matrix_data
    )

    # Convert to numeric
    for col in new_df.columns:
        new_df[col] = pd.to_numeric(new_df[col], errors="coerce")

    # Rename first columns as X and Y
    new_df.rename(columns={"V1": "X", "V2": "Y"}, inplace=True)

    # Save outputs
    matrix_out = os.path.join(output_dir, file_name.replace(".txt", "_mat.csv"))
    ions_out = os.path.join(output_dir, file_name.replace(".txt", "_ions_list.csv"))

    new_df.to_csv(matrix_out, index=False)
    ion_df.to_csv(ions_out, index=False)

    print(f"Saved: {matrix_out}")
    print(f"Saved: {ions_out}")


for file_name in files:
    create(file_name)
```
